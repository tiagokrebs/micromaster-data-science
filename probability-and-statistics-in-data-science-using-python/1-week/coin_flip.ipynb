# Flip coin - Parte 1 (Probabilidade)

```python
from numpy import random, sum
from matplotlib.pyplot import figure, hist, xlim, xlabel, ylabel, title, grid, subplot, plot
from math import sqrt
```

```python
# generate the sum of k coin flips, repat
def generate_counts(k=10000,n=100):
    X=2*(random.rand(k,n)>0.5)-1
    S=sum(X,axis=0)
    return S
```

No histograma abaixo é possivel ver que os valores tem uma divisão muito semelhante para os dois valores do teste, cara/coroa (-1/1)

```python
k=1000
n=1000
counts=generate_counts(k=k, n=n)
figure(figsize=[10,4])
hist(counts)
xlim([-k,k])
xlabel("sum")
ylabel("count")
title("Histogram of coin flip sum when flipping a fair coin %d times"%k)
grid()
```

Alterando a quantidade de jogadas múltiplas vezes a quantidade dos valores sempre é diferente porém o comportando é o mesmo, 
os valores ficam concentrados no valor 0

```python
figure(figsize=[13,3.5])
for j in range(2,5):
    k=10**j
    counts=generate_counts(k=k,n=100)
    subplot(130+j-1)
    hist(counts,bins=10)
    d=4*sqrt(k)
    plot([-d,-d],[0,30],'r')
    plot([+d,+d],[0,30],'r')
    grid()
    title('%d flips, bound=+-%6.1f'%(k,d))
```

Exibindo a escala completa das jogadas vemos que a distribuição fica cada vez mais concentrada no valor 0

```python
figure(figsize=[13,3.5])
for j in range(2,5):
    k=10**j
    counts=generate_counts(k=k,n=100)
    subplot(130+j-1)
    hist(counts,bins=10)
    xlim([-k,k])
    d=4*sqrt(k)
    plot([-d,-d],[0,30],'r')
    plot([+d,+d],[0,30],'r')
    grid()
    title('%d flips, bound=+-%6.1f'%(k,d))
```

Podemos dizer que a probabilidade dos valores resultante dessas siulações é dada por `4*sqrt(k)` onde `k` é o número de jogadas
